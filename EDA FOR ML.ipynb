```python
import ast

# Convert the string representation of the list to an actual list
merged_df['Types'] = merged_df['Types'].apply(lambda x: ast.literal_eval(x))

# Create a set to store all unique types
unique_types = set()

# Iterate over each row and update the set of unique types
for types_list in merged_df['Types']:
    unique_types.update(types_list)

# Iterate over each unique type and create a new binary column for it
for type_name in unique_types:
    merged_df[type_name] = merged_df['Types'].apply(lambda x: 1 if type_name in x else 0)

# Drop the original 'Types' column since it's no longer needed
merged_df.drop('Types', axis=1, inplace=True)

merged_df['Weighted Rating'] = merged_df['Rating'] * merged_df['Number of Ratings']
# Display the DataFrame with one-hot encoded types
merged_df

```
